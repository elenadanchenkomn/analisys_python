```python
vk['placement'] = 'vk'
inst['placement'] = 'inst'
```

```python
vk.head(1)
```

```python
df = pd.concat([vk, inst]).sort_values(['date', 'placement'])
```

```python
df.head(1)
```

```python
df = df.merge(sales, left_on=['date', 'placement'], right_on=['date', 'placement'], how='outer')
```

```python
df
```


CTR - click through rate: процент конверсии из показа в клик

```python
df['ctr'] = df['clicks'] / df.impressions
```

```python
df
```

CR1 - conversion rate 1: конверсия из клика в лид

```python
df['cr1'] = df['leads'] / df['clicks']
```

```python
df.head()
```

CPL - cost per lead, стоимость лида

```python
df['cpl'] = df['spent'] / df['leads']
```

```python
df.head(1)
```


CPQL - cost per qualified lead, стоимость квалифицированного лида

```python
df['cpql'] = df['spent'] / df['qualleads']
```

```python
df.head(1)
```

CR-Q - конверсия из лида в квалифицированного лида

```python
df['cr-q'] = df['qualleads'] / df['leads']
```

```python
df.head(5)
```


CPC - cost per click, стоимость одного клика

```python
df['cpc'] = df['spent'] / df['clicks']
```

```python
df.tail(5)
```

```python
import numpy as np

df.replace([np.inf, -np.inf], np.nan, inplace=True)
# df = df.replace([np.inf, -np.inf], np.nan)
```

```python
df.tail()
```

среднее/медианное значение CTR, CPL, CPQL по каждому рекламному каналу и среднедневной расход

```python
df2 = df.groupby('placement').agg({
    'clicks': [
        ('sum_clicks', 'sum'),
        ('mean_clicks', 'mean')
    ],
    'impressions': [
        ('sum_impressions', 'sum'),
        ('mean_impressions', 'mean')
    ],
    'spent': [
        ('total_spent', 'sum'),
        ('mean_spent', 'mean')
    ],
    'leads': [('cnt_leads', 'sum')],
    'qualleads': [('cnt_qual', 'sum')],
    'customers': [('cnt_customers', 'sum')],
    'rubs': [('sales', 'sum')]
}).reset_index()

df2.columns = df2.columns.droplevel(0)
```

```python
df2['mean_crq'] = df2['cnt_qual'] / df2['cnt_leads']
df2['mean_cpql'] = df2['total_spent'] / df2['cnt_qual']
df2['mean_cpl'] = df2['total_spent'] / df2['cnt_leads']
df2['mean_cr1'] = df2['total_spent'] / df2['sum_clicks']
df2['mean_cpc'] = df2['total_spent'] / df2['sum_clicks']
df2['cac'] = df2['total_spent'] / df2['cnt_customers']
df2['cr2'] = df2['cnt_customers'] / df2['cnt_leads']

df2['arpu'] = df2['sales'] / df2['cnt_customers']
```

```python
df2['profit'] = df2['sales'] - df2['total_spent']
```
